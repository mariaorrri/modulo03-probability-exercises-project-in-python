## Ejercicio 1 

Dos dados se lanzan una vez y se observa el total obtenido. Usa una simulación para encontrar la probabilidad estimada de que la puntuación sea mayor a 7 o un número par. Una simulación es una repetición del mismo experimento multiples veces para observar su comportamiento:

- Ejecuta el experimento 1000 veces (lanza 2 dados 1000 veces, y suma el número de ambos dados).
- Lleva ua cuenta de los números y las veces que la suma fue mayor a 7 o un número par.
- Divide el número del paso 2 entre el número de iteraciones(1000).

```python
import numpy as np
from operator import add
import random
```

```python
n_experimentos = 1000
# dado = [1 , 2 , 3 , 4 , 5 , 6]

lanzamientos_dado1 = np.zeros(n_experimentos)
lanzamientos_dado2 = np.zeros(n_experimentos)


lanzamientos_dado1 =  [random.randint( 1, 6) for _ in range(n_experimentos)]
lanzamientos_dado2 =  [random.randint( 1, 6) for _ in range(n_experimentos)]
# print(lanzamientos_dado1)
# print(lanzamientos_dado2)

# suma = list(map(add, lanzamientos_dado1, lanzamientos_dado2)) Me daba la lista como
# [np.int64(7),
#  np.int64(9),
#  np.int64(10),
#  np.int64(3),...

lista_sumados = [lanzamientos_dado1[i] + lanzamientos_dado2[i] for i in range(len(lanzamientos_dado1))]

print(lista_sumados)

mayor_7 = 0
par = 0

for valor in lista_sumados:
    if valor > 7:
        mayor_7 += 1
    if valor % 2 == 0:
        par += 1


porcentaje_mayor_7 = (mayor_7 / n_experimentos) * 100
porcentaje_par = (par / n_experimentos) * 100


print(f"Número de veces que la suma es mayor a 7: {mayor_7}")
print(f"Porcentaje de veces que la suma es mayor a 7: {porcentaje_mayor_7:}%")
print(f"Número de veces que la suma es un número par: {par}")
print(f"Porcentaje de veces que la suma es un número par: {porcentaje_par:}%")
```

## Ejercicio 2

Una caja contiene 10 bolas blancas, 20 bolas rojas y 30 bolas verdes. Si tomamos 5 bolas de la caja reemplazándolas (tomando una bola, anotando el color y luego regresandola a la caja). Queremos saber la probabilidad de:

1. Tomar 3 blancas y 2 rojas.
2. Toas todas del mismo color.

Ejecuta el experimento 1000 veces y calcula las probabilidades que mencionadas.

```python
ball_box = {}

# Crea la cajas con las bolas
for i in range(60):
    if i < 10:
        ball_box[i] = "White"
    elif (i > 9) and (i < 30):
        ball_box[i] = "Red"
    else:
        ball_box[i] = "Green"

print(ball_box)
            
colores = list(ball_box.values())
n_experimentos =10000 # Estaba de antes pero aseguramos 1000
n_bolas = 5
```

```python
# np.random.choice(colores, n_bolas)
resultados = [np.random.choice(colores, n_bolas).tolist() for _ in range(n_experimentos)]
print(resultados) 
# Hacer de otra manera, mientras vaya colocando las tiradas ir comprobando
```

```python

def contar_combinaciones(bolas, combinacion): # bolas es la lista de listas de 5 colores
    contador = 0
    for lista in bolas: # La funcion itera sobre cada sublista de la lista bolas
        if all(lista.count(color) == combinacion[color] for color in combinacion):  # Comprueba si la cominacion de colores que se quiere se cumple
            # combinacion es un diccionario con cada color y número de veces que se quiere que aparezca
            contador += 1
    return contador

def contar_todas_iguales(bolas): # bolas es la lista de listas de 5 colores
    contador = 0
    for lista in bolas: # La funcion itera sobre cada sublista de la lista bolas
        if len(set(lista)) == 1: # set(lista) comvierte la lista en un conjunto, los conjuntos son colecciones que no permiten duplicados
            # por lo que al hacerlo un set, si solo contiene el nombre de 1 color la lista va a tener un solo elemento y por tanto longitud 1
            contador += 1
    return contador

# Calcular la probabilidad de tener 3 bolas blancas y 2 rojas
combinacion_blancas_rojas = {'White': 3, 'Red': 2}
conteo_blancas_rojas = contar_combinaciones(resultados, combinacion_blancas_rojas)
probabilidad_blancas_rojas = conteo_blancas_rojas / len(resultados) *100
print(f'Probabilidad de tener 3 bolas blancas y 2 rojas: {probabilidad_blancas_rojas}%')

# Calcular la probabilidad de tener todas las bolas del mismo color
conteo_todas_iguales = contar_todas_iguales(resultados)
probabilidad_todas_iguales = conteo_todas_iguales / len(resultados)*100
print(f'Probabilidad de tener todas las bolas del mismo color: {probabilidad_todas_iguales}%')

# Comprobacion, calcula probabilidad de tener 3 bolas verdes y 2 rojas
combinacion_verdes_rojas = {'Green': 3, 'Red': 2}
conteo_verdes_rojas = contar_combinaciones(resultados, combinacion_verdes_rojas)
probabilidad_verdes_rojas = conteo_verdes_rojas / len(resultados) *100
print(f'Probabilidad de tener 3 bolas verdes y 2 rojas: {probabilidad_verdes_rojas}%')
```

La probabilidad de tener la combinación de 3 blancas y 2 rojas es muy pequeña, lo cual es lógico teniendo en cuenta que solo hay 10 bolas blancas frente a 20 rojas y 30 verdes.
Lo comprobamos igualmente con la ultima probabilidad donde pongo una combinación con más verdes y vemos que efectivamente la probailidad es mayor.
